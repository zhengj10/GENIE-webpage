```python
from glob import glob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager
```

```python
dataPath = '/Volumes/genie/GENIE webpage/iGABASnFR2/Stopped flow'
dataFiles = glob(dataPath+'/*.xlsx')
sumPath = '/Volumes/genie/GENIE webpage/iGABASnFR2/FS screening'
savePath = '/Volumes/genie/Presentations/Jihong/GENIE production/220517'
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=[6,6])
plt.rc('font', size=13) # for font size of label 1e4 in scientific style
ax = plt.subplot(111)
labels = ['602.1','602.4445','602.4409']
for i in range(3):
    dataFile = dataFiles[i]
    print(dataFile)
    df = pd.read_excel(dataFile)
    
    
    mean = np.mean([df['GABA']/df['GABA'][0],df['GABA.1']/df['GABA.1'][0],df['GABA.2']/df['GABA.2'][0],
                    df['GABA.3']/df['GABA.3'][0],df['GABA.4']/df['GABA.4'][0],df['GABA.5']/df['GABA.5'][0]],
                   axis=0)
    sem = np.std([df['GABA']/df['GABA'][0],df['GABA.1']/df['GABA.1'][0],df['GABA.2']/df['GABA.2'][0],
                    df['GABA.3']/df['GABA.3'][0],df['GABA.4']/df['GABA.4'][0],df['GABA.5']/df['GABA.5'][0]],
                   axis=0)/np.sqrt(5)
    ax.errorbar(df[df.columns[0]],mean,yerr=sem,fmt='o',label=labels[i])
    ax.set_xscale('log')
    
    
#     mean = np.mean([df['GABA'],df['GABA.1'],df['GABA.2'],df['GABA.3'],df['GABA.4'],df['GABA.5']],axis=0)
#     sem = np.std([df['GABA'],df['GABA.1'],df['GABA.2'],df['GABA.3'],df['GABA.4'],df['GABA.5']],axis=0)/np.sqrt(5)
#     ax.errorbar(df[df.columns[0]],mean,yerr=sem,fmt='o',label=labels[i])
#     ax.set_xscale('log')
    
    
    
ax.get_yaxis().set_label_coords(-0.1,0.5)
ax.set_xlabel('GABA (\u03BCM)',size=14)
# ax.set_ylabel('F',size=14)
ax.set_ylabel('F norm.',size=14)  
# Legend:
font = font_manager.FontProperties(weight='bold',style='normal',size=13)
leg = ax.legend(frameon=False,bbox_to_anchor=(0.3,1),prop=font) 
for item in leg.legendHandles: # remove line marker in legend
    item.set_visible(False)
for artist, text in zip(leg.legendHandles, leg.get_texts()): # set text the same color as line
    col = artist.get_color()
    if isinstance(col, np.ndarray):
        col = col[0]
    text.set_color(col)

plt.tick_params(labelsize=13)
plt.ticklabel_format(axis="y", style="sci", scilimits=(0,0))
plt.tight_layout()


# plt.savefig(sumPath+'/titration_raw.png', dpi=1200)
# plt.savefig(savePath+'/titration_raw.png', dpi=1200)

plt.savefig(sumPath+'/titration_norm.png', dpi=1200)
plt.savefig(savePath+'/titration_norm.png', dpi=1200)


plt.show()
```
